```python
from pwn import *
import numpy as np
```

```python
bill = [""]*50
r = remote('crypto.be.ax', 6005)
r.recvuntil(b"(y/n) ")
r.sendline(b"y")
r.recvuntil(b"New account made! Enjoy your $1.\n\n")

for i in range(50):
    r.recvuntil(b"> ")
    r.sendline(b"1")
    r.recvuntil(b"to work with: ")
    r.sendline(str(i).encode())
    r.recvuntil(b"> ")
    
    till_check = True
    while (True):
        for j in range(5):
            r.sendline(b"6")
            act6 = r.recvline().decode().split()[-1]
            r.recvuntil(b"> ")

            r.sendline(b"8")
            act8 = r.recvline().decode()
            r.recvuntil(b"> ")
            if ("Qubit successfully verified." not in act8):
                r.sendline(b"7")
                act7 = r.recvline().decode().split()[-1]
                r.recvuntil(b"> ")

                if (act7 == "+"):
                    bill[i] = "-"
                else:
                    bill[i] = "+"

                print(bill)
                r.sendline(b"9")
                till_check = False
                break
                
        if (not till_check):
            break

        for j in range(5):
            r.sendline(b"7")
            act7 = r.recvline().decode().split()[-1]
            r.recvuntil(b"> ")

            r.sendline(b"8")
            act8 = r.recvline().decode()
            r.recvuntil(b"> ")
            if ("Qubit successfully verified." not in act8):
                r.sendline(b"6")
                act6 = r.recvline().decode().split()[-1]
                r.recvuntil(b"> ")

                if (act6 == "1"):
                    bill[i] = "0"
                else:
                    bill[i] = "1"

                print(bill)
                r.sendline(b"9")
                till_check = False
                break
                
        if (not till_check):
            break
    
r.recvuntil(b"> ")
r.sendline(b"2")
r.recvline()
print("".join(bill).encode())
r.sendline("".join(bill).encode())
print(r.recvline())
```
